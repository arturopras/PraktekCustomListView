# 2. Text Preprocessing

```python
#Cleaning Data
def data_clean(text):
  text = re.sub(r'@[A-Za-z0-9_]+', '', text)
  text = re.sub(r'#\w+', '', text)
  text = re.sub(r'RT[\s]+', '', text)
  text = re.sub(r'https?://\S+', '', text)

  text = re.sub(r'[^A-Za-z0-9 ]', '', text)
  text = re.sub(r'\s+', ' ', text).strip()

  return text

df['clean_review'] = df['review'].apply(data_clean)
```

```python
#Hasil Cleaning
df.head()
```

```python
#Case Folding
df['case_review'] = df['clean_review'].str.lower()
df.head()
```

```python
#StopWord
import Sastrawi
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
more_stopword ={}

stop_words = StopWordRemoverFactory().get_stop_words()
new_array = ArrayDictionary(stop_words)
stop_words_remover_new = StopWordRemover(new_array)

def stop_words(str_text):
  str_text = stop_words_remover_new.remove(str_text)
  return str_text

df['stop_review'] = df['case_review'].apply(lambda x: stop_words(x))
df.head()
```

```python
#Tokenisasi
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')

df['token_review'] = df['stop_review'].apply(word_tokenize)
```

```python
df.head()
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def stemming(Review):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  do = []
  for w in Review:
    dt = stemmer.stem(w)
    do.append(dt)
  d_clean = []
  d_clean = " ".join(do)
  print(d_clean)
  return d_clean

df['stem_review'] = df['token_review'].apply(stemming)
```

```python
#menghapus baris hilang
df = df[df['stem_review'].apply(lambda x: len(x) > 0)]
```

# 4. Data Preparation

```python
#Memisahkan Data Latih dan Data Uji
x = df_cleaned['stem_review']
y = df_cleaned['sentimen']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print("Data latih: ",len(x_train))
print("Data uji: ",len(x_test))
```

```python
#Menggunakan CountVectorizer untuk mengubah teks menjadi fitur numerik
vectorizer = CountVectorizer()
x_train_vec = vectorizer.fit_transform(x_train)
x_test_vec = vectorizer.transform(x_test)
```

```python
#Melakukan oversampling menggunakan SMOTE pada data latih
smote = SMOTE(random_state=42)
x_train_res, y_train_res = smote.fit_resample(x_train_vec, y_train)
```

```python
#Contoh data hasil sampling
sentimen_counts = y_train_res.value_counts()
plt.figure(figsize=(5, 3))
plt.bar(sentimen_counts.index, sentimen_counts.values, color={"red", "lightskyblue"})
plt.title('Visualisasi Sentimen menggunakan SMOTE')
plt.xlabel('Sentimen')
plt.ylabel('Jumlah')
plt.xticks(sentimen_counts.index, ['0', '1'])
plt.show()
```
